```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('files/Churn_Modelling.csv')
```

```python
# most people are from france
plt.figure(figsize=(15, 8))
sns.countplot(x=data.Geography)
```

```python
# most are male
plt.figure(figsize=(15, 8))
sns.countplot(x=data.Gender)
```

```python
plt.figure(figsize=(15, 8))
sns.heatmap(data=data.corr(), annot=True, fmt='.1g')
```

```python
plt.figure(figsize=(15, 8))
sns.countplot(x=data.Tenure)
plt.title('Count of each tenure', fontsize=15)
plt.xlabel('Tenure', fontsize=15)
plt.ylabel('Count', fontsize=15)
```
